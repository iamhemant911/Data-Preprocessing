```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data
```

# Principal Component Analysis

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
x_pca
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```

# t-SNE

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=2000,random_state = 17)

df_tsne_scaled = tsne.fit_transform(scaled_data)

plt.figure(figsize=(12,10))
plt.scatter(df_tsne_scaled[:, 0], df_tsne_scaled[:, 1], c=cancer['target'], 
            alpha=0.7, s=40)
plt.colorbar()
plt.title('t-SNE projection');
```

# Locally Linear Embedding

```python
from sklearn.manifold import LocallyLinearEmbedding
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
lle_scaled = lle.fit_transform(scaled_data)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(lle_scaled[:,0],lle_scaled[:,1],c=cancer['target'])
plt.title('lle projection')
```
